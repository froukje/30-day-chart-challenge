# Setup

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
from matplotlib import cm
```

# Data

```python
data = {"total": 18547784, 
        "Common carp": 4641275, 
        "Rainbow trout": 6158172, 
        "Salmon-colored Rainbow trout": 1627341, 
        "Alsatian char": 1585572, 
        "Brown trout": 618043,
        "European eel": 1202246,
        "North African catfish": 1193137}

data_percent = {k: v for k, v in sorted(data_percent.items(), key=lambda item: item[1], reverse=True)}
data_percent
```

```python
fig = plt.figure(
    FigureClass=Waffle,
    figsize = (10,10),
    rows=5,
    #columns=20,  
    values=data_percent,
    title={'label': 'Fish Production Volume produced in 2019', 'loc': 'left'},
    labels=[f"{k} ({v}%)" for k, v in data_percent.items()],
    #legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': int(len(data)/2), 'framealpha': 0},
    legend={'loc': 'upper left', 'bbox_to_anchor': (1.05, 1.05)},
    icons='fish',
    icon_legend=True,
    font_size=24,
    tight=False,
    vertical=True,
    block_aspect_ratio=1,
    #colors=['#F5B7B1','#E8DAEF','#AED6F1', '#A2D9CE','#F9E79F','#F5CBA7','#F9E79F']
    colors=['#A569BD','#5DADE2', '#45B39D','#F4D03F','#F5B041','#EC7063','pink']
)
fig.set_facecolor('#EBF5FB')
fig.set_size_inches(10, 10)
plt.savefig("../plots/day02_pictogram.png", facecolor='#EBF5FB', bbox_inches='tight',pad_inches=.5)
```

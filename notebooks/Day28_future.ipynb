# Read Data

# Plot

```python
# possible impacts per year
years = list(df_ast["Period End"] - df_ast["Period Start"])
years_start = list(df_ast["Period Start"])
years_end = list(df_ast["Period End"])
impacts_per_year =  list(df_ast["Possible Impacts"]/years)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,5))

s = [years[n]/2 for n in range(len(years))]
ax.scatter(years_start, impacts_per_year, c='#20497C', marker='o', s=s)
ax.set_xlim(2015,2190)
ax.set_title("Possible asteroid impacts from Startyear on")
ax.set_ylabel("Possible number of impacts")
ax.text(x=2030, y=10, s=f"Each dot represents one asteroid, \nthe size is proportional to the number of years predicted", color='#20497C');
plt.savefig("../plots/day28_future.png");
```

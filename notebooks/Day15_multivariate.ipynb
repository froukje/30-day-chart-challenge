# Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import EllipseCollection
import seaborn as sns
sns.set_theme(style="ticks")
```

# Read Data

```python
df_penguins = pd.read_csv("../data/penguins/penguins.csv")
df_penguins.head()
```

# Correlation Plots

```python
df_penguins = df_penguins[["species", "island", "bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g", "sex"]]
```

```python
plot = sns.pairplot(df_penguins, hue="species", palette={"Adelie": "#40E0D0", "Gentoo": "#6495ED", "Chinstrap": "#CCCCFF"})
fig = plot.fig
fig.subplots_adjust(top=0.93, wspace=0.3)
fig.suptitle("Penguins Pairplots")
plt.savefig("../plots/day15_multivariate.png");
```

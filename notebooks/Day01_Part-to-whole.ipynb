# Prepare Data for Plotting

* Define two new dataframes: catches per fish family per year (df_families) and catches per species per year (df_names)
* Find the 5 most catched fish families / species over the entire time period
* Get the percentage of these families / species per year

```python
# dictionary of familiy names and their time series
years = ["2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014"]
Fam = {}
for year in years:
    for fam in df_final["Family"]:
        species = df_final[df_final["Family"] == fam].index
        if fam not in Fam.keys():
            Fam[fam] = {}
        Fam[fam][year] =  sum([int(df_final.loc[species[j]][year]) for j in range(len(species))])
#Fam
```

```python
Fam.keys()
```

```python
# dictionoray of species (english name) and their time series
years = ["2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014"]
EnName = {}
for year in years:
    for name in df_final["English_name"]:
        species = df_final[df_final["English_name"] == name].index
        if name not in EnName.keys():
            EnName[name] = {}
        EnName[name][year] =  int(df_final.loc[species][year])
```

```python
EnName.keys()
```

```python
# save families as dataframe
df_families = pd.DataFrame() 
for key in Fam.keys():
    df_families[key] = [Fam[key][year] for year in years]
df_families["year"] = years
df_families = df_families.set_index("year", drop=True)
df_families
```

```python
# save species (english name) as dataframe
df_names = pd.DataFrame() 
for key in EnName.keys():
    df_names[key] = [EnName[key][year] for year in years]
df_names["year"] = years
df_names = df_names.set_index("year", drop=True)
df_names
```

```python
# 5 most fish families in dataframe
sum_of_allyears = df_families.cumsum().loc["2014"]
five_highest = sum_of_allyears.sort_values()[::-1][:5]
print(f"The 5 most catched fish families in 2006 to 2014 are: \n{five_highest}")
```

```python
# 5 most fish species (english name) in dataframe
sum_of_allyears_names = df_names.cumsum().loc["2014"]
five_highest_names = sum_of_allyears_names.sort_values()[::-1][:5]
print(f"The 5 most catched fishs in 2006 to 2014 are: \n{five_highest_names}")
```

```python
# total catch per year
total_years = df_families.sum(axis=1)
```

```python
# evolution of 5 highest over time (in percentage)
five_families = ["Clupeidae", "Gadidae", "Scombridae", "Carangidae", "Crangonidae"]
catches = {}
for year in years:
    for family in five_families:
        if family not in catches.keys():
            catches[family] = [df_families.loc[year][family] / total_years[year]]
        else:
            catches[family].append(df_families.loc[year][family] / total_years[year])
#catches
```

```python
# evolution of 5 highest species (english names)over time (in percentage)
five_names = ["Atlantic herring", "European sprat", "Atlantic mackerel", "Common shrimp", "Blue whiting(=Poutassou)"]
catches_names = {}
for year in years:
    for name in five_names:
        if name not in catches_names.keys():
            catches_names[name] = [df_names.loc[year][name] / total_years[year]]
        else:
            catches_names[name].append(df_names.loc[year][name] / total_years[year])
#catches_names
```

```python
# save families as dataframe (in percentage)
df_five = pd.DataFrame(catches)
df_five["year"] = years
df_five = df_five.set_index("year", drop=True)
df_five["other"] = 1 - df_five.sum(axis=1)
df_five = round(df_five*100,2)
df_five
```

```python
# save names as dataframe (in percentage)
df_five_names = pd.DataFrame(catches_names)
df_five_names["year"] = years
df_five_names = df_five_names.set_index("year", drop=True)
df_five_names["other"] = 1 - df_five_names.sum(axis=1)
df_five_names = round(df_five_names*100,2)
df_five_names
```

# Final Plotting 1-whole: Choose stacked barplot

1. Fish catched in Germany by Family
1. Fish catched in Germany by Specie

```python
# plot evolution over time of five highest families
ax = df_five.plot(
    kind = 'barh',
    stacked = True,
    title = 'Stacked Bar Graph',
    color=("#F1948A", "#C39BD3", "#85C1E9", "#82E0AA", "#F8C471", "#E59866"),
    figsize=(15,8),
    mark_right = True);
# .patches is everything inside of the chart, lines and
# rectangles and circles and stuff. In this case we only
# have rectangles!
for rect in ax.patches:
    # Find where everything is located
    height = rect.get_height()
    width = round(rect.get_width(),2)
    x = rect.get_x()
    y = rect.get_y()
    
    # The width of the bar is the percentige
    label_text = f'{width}%'
    
    # ax.text(x, y, text)
    label_x = x + width / 2
    label_y = y + height / 2
    ax.text(label_x, label_y, label_text, ha='center', va='center')
    ax.legend(labels=df_five.columns,fontsize= 14, bbox_to_anchor=(1.02, 1.03), loc='upper left')
    ax.set_title("Fish Catches in Germany 2006 to 2014 (in %)",fontsize= 16)
    ax.set_xlabel("Percent");
```

```python
# plot evolution over time of five highest species
ax = df_five_names.plot(
    kind = 'barh',
    stacked = True,
    title = 'Stacked Bar Graph',
    figsize=(15,8),
    color=("#F1948A", "#C39BD3", "#85C1E9", "#82E0AA", "#F8C471", "#E59866"),
    mark_right = True);
# .patches is everything inside of the chart, lines and
# rectangles and circles and stuff. In this case we only
# have rectangles!
for rect in ax.patches:
    # Find where everything is located
    height = rect.get_height()
    width = round(rect.get_width(),2)
    x = rect.get_x()
    y = rect.get_y()
    
    # The width of the bar is the percentige
    label_text = f'{width}%'
    
    # ax.text(x, y, text)
    label_x = x + width / 2
    label_y = y + height / 2
    ax.text(label_x, label_y, label_text, ha='center', va='center')
    lgd = ax.legend(labels=df_five_names.columns,fontsize= 14, bbox_to_anchor=(1.02, 1.015), loc='upper left')
    ax.set_title("Fish Catches in Germany 2006 to 2014 (in %)",fontsize= 16)
    ax.set_xlabel("Percent");
plt.savefig("../plots/day1_1-to-whole.png", bbox_extra_artists=(lgd,), bbox_inches='tight')
```

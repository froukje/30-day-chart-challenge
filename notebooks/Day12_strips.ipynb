# Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read Data

```python
df = pd.read_csv("../data/covid19/country_vaccinations.csv")
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df_ger = df[df["country"] == "Germany"]
df_ger
```

```python
df_ger["date"]
```

```python
group1 = df_ger["people_vaccinated_per_hundred"]
group2 = 100 - group1
```

```python
group1
```

```python
group2
```

```python
fig, ax = plt.subplots(1,1, figsize=(25,10))

p1 = ax.bar(np.arange(len(group1)), group1, width=.4, label='vaccinated', color="darkgreen")
p2 = ax.bar(np.arange(len(group1)), group2, width=.4,
            bottom=group1, label='rest', color="lightgreen")

ax.set_ylabel('people vaccinated per hundred')
ax.set_title('People vaccinated per hundred in Germany', fontsize=16, pad=20)
ax.set_xticks(range(len(df_ger["date"])))
ax.set_xticklabels(df_ger["date"], rotation='vertical')
ax.legend()
plt.savefig("../plots/day12_stripes.png")

plt.show()
```

# Read Data

```python
# plot different sceanrios for "Population"
ts_main_series = data[1]
ts_low_immigration = data[3]
ts_high_immigration = data[5]
ts_zero_immigration = data[7]
```

# Plot

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(data[0], ts_main_series, color='#138D75')
ax.plot(data[0], ts_low_immigration, color='#D98880')
ax.plot(data[0], ts_high_immigration, color='#7B241C')
ax.plot(data[0], ts_zero_immigration, color='#C0392B')
ax.legend(['main series', 'low immigration', 'high immigration', 'zero immigration'])
ax.set_title("Immigration Projections - Alternative Projections for the United States: 2017-2060")
plt.savefig("../plots/day25_demographic.png");
```

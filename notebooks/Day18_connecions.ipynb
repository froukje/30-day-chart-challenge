# Setup

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

# Plot

```python
fig, ax = plt.subplots(1, 1, figsize=(10,15))
#ax = plt.axes(projection=ccrs.InterruptedGoodeHomolosine())
ax = plt.axes(projection=ccrs.Robinson(central_longitude=0))
ax.stock_img()

hh_lon, hh_lat = 9.99, 53.55
san_lon, san_lat = -70.67, -33.45
an_lon, an_lat = -149.86, 61.22
kap_lon, kap_lat = 18.42, -33.92 
tok_lon, tok_lat = 139.84, 35.65
hob_lon, hob_lat = 147.32, -42.88
osl_lon, osl_lat = 10.76, 59.91
mex_lon, mex_lat = -99.13, 19.43

plt.plot([san_lon, hh_lon], [san_lat, hh_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([san_lon, hh_lon], [san_lat, hh_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.text(san_lon - 3, san_lat - 12, 'Santiago',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.text(hh_lon + 3, hh_lat - 12, 'Hamburg',
         horizontalalignment='left',
         transform=ccrs.Geodetic())

plt.plot([an_lon, hh_lon], [an_lat, hh_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([an_lon, hh_lon], [an_lat, hh_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.text(an_lon - 3, an_lat - 12, 'Anchorage',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.plot([kap_lon, hh_lon], [kap_lat, hh_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([kap_lon, hh_lon], [kap_lat, hh_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.text(kap_lon - 3, kap_lat - 12, 'Cape Town',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.plot([tok_lon, hh_lon], [tok_lat, hh_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([tok_lon, hh_lon], [tok_lat, hh_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.text(tok_lon - 3, tok_lat - 12, 'Tokyo',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.plot([hob_lon, osl_lon], [hob_lat, osl_lat],
         color='red', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([hob_lon, osl_lon], [hob_lat, osl_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.text(hob_lon + 3, hob_lat - 12, 'Hobart',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.text(osl_lon + 3, osl_lat + 8, 'Oslo',
         horizontalalignment='left',
         transform=ccrs.Geodetic())

plt.plot([mex_lon, osl_lon], [mex_lat, osl_lat],
         color='red', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([mex_lon, osl_lon], [mex_lat, osl_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.text(mex_lon + 3, mex_lat - 12, 'Mexiko City',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.text(osl_lon + 3, osl_lat + 8, 'Oslo',
         horizontalalignment='left',
         transform=ccrs.Geodetic())

plt.plot([hob_lon, mex_lon], [hob_lat, mex_lat],
         color='red', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([hob_lon, mex_lon], [hob_lat, mex_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )
ax.set_title("Distances: straight lines vs. geodetic")
plt.savefig("../plots/day18_connections.png", bbox_inches = 'tight')
plt.show()
```

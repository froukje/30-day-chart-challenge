# Setup

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
# import distributions
from scipy.stats import uniform, norm, gamma, expon, poisson, binom
```

# Prepare Data

```python
data = []
```

```python
# uniform distribution
data.append(uniform.rvs(size=10000, loc = 10, scale=20))
# normal distribution
data.append(norm.rvs(size=10000,loc=0,scale=1))
# gamma distribution
data.append(gamma.rvs(a=5, size=10000))
# exponetial distribution
data.append(expon.rvs(scale=1,loc=0,size=1000))
# poisson distribution
data.append(poisson.rvs(mu=3, size=10000))
# binomial distribution
data.append(binom.rvs(n=10,p=0.8,size=10000))
```

# Plot

```python
title = ["Uniform Distribution", "Normal Distribution", "Gamma Distribution", 
         "Exponetial Distribution", "Poisson Distribution", "Binomial Distribution"]
```

```python
fig, axes = plt.subplots(3,2, figsize=(15,10),)
layout_tight = True
plt.subplots_adjust(hspace = 0.5)
for i, ax in enumerate(axes.flatten()):
    ax = sns.histplot(data[i],
                  bins=50,
                  color='#006699', alpha=.5,
                  ax=ax, element="step")
    #ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
    ax.set_title(title[i]);
plt.savefig("../plots/day09_statistics.png")
```
